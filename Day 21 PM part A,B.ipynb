```
# regression
X = np.arange(10)
y_reg = 2 * X + 3

# classification
y_clf = [1 if x > 5 else 0 for x in X]
```

```
import matplotlib.pyplot as plt

plt.scatter(X, y_reg)
plt.title("Regression")
plt.show()

plt.scatter(X, y_clf)
plt.title("Classification")
plt.show()
```

```
def predict(x):
    return 2*x + 3

y_pred = [predict(x) for x in X]

mse = sum((y_reg - y_pred) ** 2) / len(X)
print(mse)
```

```
pred = [1 if x > 5 else 0 for x in X]

correct = 0
for i in range(len(X)):
    if pred[i] == y_clf[i]:
        correct += 1

accuracy = correct / len(X)
print(accuracy)
```
